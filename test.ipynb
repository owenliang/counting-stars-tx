```python
import json 
# 评测函数
def run_test_case(model_fn,model_name,test_cases):
    stats=[]
    for sample in test_cases:
        resp=model_fn(model_name,sample['prompt'])
        resp=resp.strip('`')
        if resp.startswith('json'):
            resp=resp[4:]
        pred_stars=json.loads(resp)
        pred_stars=set(pred_stars['小企鹅'])

        
        if sample['test_type']=='acquisition':
            for a in sample['acquisition_stars']:
                if a in pred_stars:
                    score=1
                else:
                    score=0
                stats.append(score)
        else:
            for a,r in zip(sample['acquisition_stars'],sample['reasoning_stars']):
                if a in pred_stars and r not in pred_stars:
                    score=1
                elif a in pred_stars and r in pred_stars:
                    score=0.5
                elif a not in pred_stars and r in pred_stars:
                    score=0.25
                else:
                    score=0
                stats.append(score)
    print('acc:%.2f'%(sum(stats)/len(stats)))
```
